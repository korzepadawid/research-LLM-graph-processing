```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import SpectralClustering
```

```python
edges = [
    (1, 2), (2, 3), (3, 4), (3, 5), (4, 6), (5, 6), (6, 7), (7, 8),
    (7, 9), (8, 10), (9, 10), (6, 12), (12, 13), (13, 14), (14, 15)
]
G = nx.Graph()
G.add_edges_from(edges)
```

```python

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, cmap=plt.cm.Set3)
plt.show()
```

# Spectral clustering

```python
adj_matrix = nx.to_numpy_array(G)

clustered = SpectralClustering(n_clusters=5, affinity='precomputed').fit_predict(adj_matrix)

colors = [f"C{label}" for label in clustered]
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color=colors, cmap=plt.cm.Set3, node_size=800, font_size=10)
plt.title("Graph Colored by Spectral Clustering")
plt.show()
```
